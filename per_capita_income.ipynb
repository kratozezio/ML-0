```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("datasets/canada_per_capita_income.csv")
```

```python
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.scatter(data['year'], data['per capita income (US$)'])
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(data['year'], data['per capita income (US$)'], test_size = 0.2)
```

```python
plt.scatter(x_train, y_train, label="Training Data", color='b', alpha=0.7)
plt.scatter(x_test, y_test, label="Testing Data", color='r', alpha=0.7)
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.legend()
plt.title("Train vs Test Split")
plt.show()
```

```python
model = LinearRegression()
model.fit(x_train.values.reshape(-1, 1), y_train.values)
```

```python
prediction = model.predict(x_test.values.reshape(-1, 1))
```

```python
plt.plot(x_test.values, prediction, label='Linear Regression Line', color='b')
plt.scatter(x_test.values, y_test.values, label='Testing data', color='r')
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.legend()
plt.title("Prediction Line vs Test Data")
plt.show()
```

```python
# Constants

score = model.score(x_test.values.reshape(-1, 1), y_test.values)
slope = model.coef_
intercept = model.intercept_
print(f"Intercept: {intercept}")
print(f"Slope: {slope}")
print(f"score: {score}")
```
