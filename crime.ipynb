```python
from pydataset import data
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Load the dataset

dataset = data('Crime')
```

```python
# Plot the data

dataset.plot(kind='scatter', x='density', y='crmrte')
plt.show()
```

```python
# Split training and testing data

X_train, X_test, Y_train, Y_test = train_test_split(dataset.density, dataset.crmrte)
plt.scatter(X_train, Y_train, label='Training Data', color='r', alpha=0.7)
plt.scatter(X_test, Y_test, label='Testing Data', color='b', alpha=0.7)
plt.title("Train Test Split")
plt.legend()
plt.show()
```

```python
# Build and train the model

model = LinearRegression()
model.fit(X_train.values.reshape(-1, 1), Y_train)
```

```python
 # Predictions

 predictions = model.predict(X_test.values.reshape(-1, 1))
 plt.plot(X_test, predictions, label="Prediction Line", color='b')
 plt.scatter(X_test, Y_test, label='Testing data', color='r')
 plt.title("Prediction vs Testing Data")
 plt.legend()
 plt.show()
```

```python
# Score the model

slope = model.coef_
intercept = model.intercept_
score = model.score(X_test.values.reshape(-1, 1), Y_test.values)
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"Score (R^2): {score}")
```
