```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

#### Read a csv file

```python
df = pd.read_csv("datasets/avocado.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

#### View first/last few items

```python
df.head(3)
```

```python
df.tail()
```

##### Conditionals

```python
albany_df = df[ df['region'] == "Albany" ]
albany_df.head()
```

###### Change Index

```python
# same as albany_df = albany_df.set_index("Date")
albany_df.set_index("Date", inplace=True)
albany_df.head()
```

```python
albany_df.plot()
```

```python
albany_df['Total Volume'].plot()
```

```python
# To smoothen graph
albany_df['Total Volume'].sort_index(inplace=True)
albany_df['Total Volume'].rolling(25).mean().plot()
```

```python
albany_df.head()
```

###### Create a copy of a dataframe

```python
# df.copy() returns a copy of dataframe df

albany_df = df.copy()[ df['region'] == "Albany" ]
albany_df['Date'] = pd.to_datetime(albany_df['Date'])
albany_df.set_index('Date', inplace=True)
albany_df.head()
```
