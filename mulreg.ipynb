```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
dataset = df = pd.read_csv('datasets/CarPrice_Assignment.csv')
plt.scatter(dataset['carwidth'], dataset['price'])
plt.title("Car Width vs Price")
plt.xlabel("Car Width")
plt.ylabel("Price")
plt.plot()

# Linear relationship
```

```python
plt.scatter(dataset['horsepower'], dataset['price'])
plt.title("HP vs Price")
plt.xlabel("Horsepower")
plt.ylabel("Price")
plt.plot()

# Linear relationship
```

```python
x = dataset[['carwidth', 'horsepower']]
y = dataset['price']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
# plt.scatter(x_train['carwidth'], y_train, color='b', alpha=0.7)
# plt.scatter(x_train['horsepower'], y_train, color='r', alpha=0.7)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
np.shape(predictions)
```

```python
model.predict([[67.9, 160]])
```

```python
model.score(x_test, y_test)
```

```python
print(model.coef_)
print(model.intercept_)
```
