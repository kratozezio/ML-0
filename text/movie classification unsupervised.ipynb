```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("../datasets/Movies_Dataset.csv", encoding='utf-8')
# df = df.drop(['title', 'id'], axis=1)
df.head()
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['overview'].values.astype('U'))
# count_vect = vectorizer.fit_transform(["This is my first try try first."])
```

```python
# words = vectorizer.get_feature_names_out()
# count = count_vect.toarray().flatten()
# for word, count in zip(words, count):
    # print(f"{word}: {count}")
```

```python
model = KMeans(n_clusters=20, max_iter=100)
model.fit(X)
```

```python
df['cluster'] = model.labels_
df.head()
```

```python
cluster_centers = model.cluster_centers_.argsort()[:, ::-1]
features = vectorizer.get_feature_names_out()
```

```python
for i in range(20):
    print(f"Cluster id: {i}")
    for j in cluster_centers[i, :10]:
        print(f"{features[j]}")
    print("-----------")
```
